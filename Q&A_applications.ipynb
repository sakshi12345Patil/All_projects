```python
import random
import time

def Addition(x, y):
    return x + y

def Subtraction(x, y):
    return x - y

def Multiplication(x, y):
    return x * y

def Greaterthan(x, y):
    return x > y

def Lessthan(x, y):
    return x < y

def EqualsTo(x, y):
    return x == y

def NotEqualsTo(x, y):
    return x != y

operators = ["+", "-", "x", ">", "<", "==", "!="]
operands = list(range(1, 11))
player_scores = {1: 0, 2: 0}
player_times = {1: 0, 2: 0}

for player in range(1, 3):
    print("\n\nPlayer", player, "play your Turn")
    print("Press ENTER when you are ready")
    input()
    score = 0
    total_time = 0
    for question in range(1, 6):
        a = random.choice(operands)
        b = random.choice(operands)
        op = random.choice(operators)
        print("Question", question, " ", a, op, b, end=": ")

        if op == "+":
            correct_ans = Addition(a, b)
        elif op == "-":
            correct_ans = Subtraction(a, b)
        elif op == "x":
            correct_ans = Multiplication(a, b)
        elif op == ">":
            correct_ans = Greaterthan(a, b)
        elif op == "<":
            correct_ans = Lessthan(a, b)
        elif op == "==":
            correct_ans = EqualsTo(a, b)
        elif op == "!=":
            correct_ans = NotEqualsTo(a, b)

        start_time = time.time()
        ans = input().strip()
        end_time = time.time()
        time_taken = end_time - start_time
        total_time += time_taken

        if op in [">", "<", "==", "!="]:
            ans = ans.lower() in ["true", "1"]
        else:
            try:
                ans = int(ans)
            except ValueError:
                ans = None

        if ans == correct_ans:
            print("Correct!")
            score += 1
        else:
            print("Wrong! The correct answer was", correct_ans)

    print("Score for Player", player, "is:", score)
    print("Total time taken by Player", player, "is:", round(total_time, 2), "seconds")
    player_scores[player] = score
    player_times[player] = total_time

if player_scores[1] > player_scores[2]:
    print("Player 1 wins!")
elif player_scores[1] < player_scores[2]:
    print("Player 2 wins!")
else:
    print("It's a tie!")

if player_times[1] < player_times[2]:
    print("Player 1 was faster!")
elif player_times[1] > player_times[2]:
    print("Player 2 was faster!")
else:
    print("Both players took the same amount of time!")
```
