```python
#---- Library Project ----

#* Create file "all_books.txt"
 # book number, book title, book author, book publication

#* Create file "all_students.txt"
 # enrollment number, name, email, mobile number, class

#* Create file "all_issued.txt"
 # enrollment number, book number, issue date, return date

#=============================================

#* Operations:
#	Issue Book
#	Return Book
#	View Not Returned Books
#	Add New Book
#	Add New Student
#	Search Book
#	Search Student
#	Student History
#	Book History

#==================================================

#import datetime

#a = datetime.date.today()
#idate = str(a.day) + "-" + str(a.month) + "-" + str(a.year)
#print(idate)

#==============================


import datetime

def get_date():
  a = datetime.date.today()
  cdate = str(a.day) + "-" + str(a.month) + "-" + str(a.year)
  return cdate

def issue_book():
  # accept enrollment number of borrower, book number to
  # borrow, idate, NA
  # write this into file "all_issued.txt"
  enr = input("Enter Enrollment Number of Borrower : ")
  bnum = input("Enter Book Number to Issue : ")
  idate = get_date()

  fobj = open("all_issued.txt", "a")
  fobj.write(enr + "," + bnum + "," + idate + "," + "NR\n")
  fobj.close()
  print("Book Issued...")
  input()

def return_book():
  # accept book number to return (which student brought)
  # simply replace "NA" with rdate.
  bnum = input("Enter Book Number to Return : ")

  fobj = open("all_issued.txt", "r")
  all_data = fobj.readlines()
  fobj.close()

  rdate = get_date()
  ind = 0
  while ind<len(all_data):
    oneline = all_data[ind]
    ls = oneline.split(",")
    if ls[1]==bnum and ls[3]=="NR\n":
      ls[3] = rdate + "\n"
      newline = ",".join(ls)
      all_data[ind] = newline
      break
    ind = ind + 1

  fobj = open("all_issued.txt", "w")
  fobj.writelines(all_data)
  fobj.close()
  print("Book Returned...")
  input()

def view_not_returned_books():
  # print list of books and enrollemnt number which are
  # yet not returned.
  # refer file "all_issued.txt"

  fobj = open("all_issued.txt", "r")
  all_data = fobj.readlines()
  fobj.close()

  for oneline in all_data:
    ls = oneline.split(",")
    if ls[3] == "NR\n":
      print(ls[0] + " || " + ls[1] + " || " + ls[2])
  input()

def add_new_student():
  # accept enrollment number, name, email, mobile and class
  # number of new student and write it in "all_students.txt"
  enr = input("Enter Enrollment Number: ")
  name = input("Enter Name: ")
  email = input("Enter Email: ")
  mobile = input("Enter Mobile Number: ")
  class_name = input("Enter Class: ")

  with open("all_students.txt", "a") as f:
        f.write(f"{enr},{name},{email},{mobile},{class_name}\n")

  print("Student added successfully!")
  input()

def add_new_book():
  # accept book number, book title, author name and
  # publication name of new book and write it in
  # file "all_books.txt"
    book_num = input("Enter Book Number: ")
    title = input("Enter Book Title: ")
    author = input("Enter Author Name: ")
    publication = input("Enter Publication: ")

    with open("all_books.txt", "a") as f:
        f.write(f"{book_num},{title},{author},{publication}\n")

    print("Book added successfully!")
    input()

def search_book():
  # accept book number and search and print it's title,
  # author name and publication. Refer "all_books.txt"

    bnum = input("Enter Book Number to search: ")
    found = False

    with open("all_books.txt", "r") as f:
        for line in f:
            ls = line.strip().split(",")
            if ls[0] == bnum:
                print("Book Title   :", ls[1])
                print("Author       :", ls[2])
                print("Publication  :", ls[3])
                found = True
                break

    if not found:
        print("Book not found.")
    input()

def search_student():
  # accept enrollment number and search and print name,
  # class, email and mobile number of that student.
  # Refer "all_students.txt"
  enum=input("Enter Enrollment Number : ")
  found=False

  with open("all_students.txt","r") as f:
    for line in f:
      ls=line.strip().split(",")
      if ls[0]==enum:
        print("Name : ",ls[1])
        print("Class : ",ls[4])
        print("Email : ",ls[2])
        print("Mobile : ",ls[3])
        found=True
        break
  if not found:
    print("Student not found.")
  input()

def student_history():
  # accept enrollment number and print his/her complete
  # history of books he/she borrowed till date.
  # This will helpful to librarian to check whether any
  # book is still pending with that student or not.
  # Refer "all_issued.txt"
  enr = input("Enter Student Enrollment Number : ")

  fobj = open("all_issued.txt", "r")
  all_data = fobj.readlines()
  fobj.close()

  for oneline in all_data:
    ls = oneline.split(",")
    if ls[0] == enr:
      print(ls[1] + " || " + ls[2] + " || " + ls[3], end="")
  input()

def book_history():
  # accept book number and print history of this book.
  # This will helpful to librarian to place new order
  # of that book.
  bnum = input("Enter Book Number : ")
  fobj = open("all_issued.txt", "r")
  all_data = fobj.readlines()
  fobj.close()

  for oneline in all_data:
    ls = oneline.split(",")
    if ls[1] == bnum:
      print(ls[0] + " || " + ls[2] + " || " + ls[3], end="")
  input()

while True:
  print("Select operation")
  print("1 - Issue Book")
  print("2 - Return Book")
  print("3 - View Not Returned Books")
  print("4 - Search Student")
  print("5 - Search Book")
  print("6 - Add New Student")
  print("7 - Add New Book")
  print("8 - Student History")
  print("9 - Book History")
  print("0 - Exit")
  ch = int(input("Provide your choice : "))
  if ch==1: issue_book()
  elif ch==2: return_book()
  elif ch==3: view_not_returned_books()
  elif ch==4: search_student()
  elif ch==5: search_book()
  elif ch==6: add_new_student()
  elif ch==7: add_new_book()
  elif ch==8: student_history()
  elif ch==9: book_history()
  elif ch==0:
    print("ok.. Bye!")
    break
  else:
    print("Invalid choice! Please select a valid option.")

```
