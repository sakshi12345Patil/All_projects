```python
# snake ladder game (sap-sidi)
import random

print("Welcome to Snake Ladder Game")
snakes = {32: 10, 36: 6, 48: 26, 62: 18, 88: 24, 95: 56, 97:78}
ladders = {1:38 ,4: 14, 8: 30, 21: 42, 28: 76, 50: 67, 71: 92, 80: 99}

player1 = input("Enter the name of player 1 : ")
player2 = input("Enter the name of player 2 : ")

dice = [1, 2, 3, 4, 5, 6]

player1_pos = 0
player2_pos = 0

player1_turn = True
player2_turn = False
while True:
    if player1_turn:
        print(f"its {player1}'s turn")
        input("Press enter to roll the dice")
        dice_value = random.choice(dice)
        print(f"{player1} got {dice_value}")
        p1_new_position = player1_pos + dice_value

        if p1_new_position >= 100:
            print("oopps....try again in next turn")


        elif p1_new_position in snakes:
            print("Oops! Snake Bite ")
            p1_new_position = snakes[p1_new_position]
            print(f"your new position is {p1_new_position}")
        elif p1_new_position in ladders:
            print("wow! You got a ladder")
            p1_new_position = ladders[p1_new_position]
            print(f"your new position is {p1_new_position}")
        else:
            player1_pos = p1_new_position
            print(f"your new position is {player1_pos}")
            if player1_pos == 100:
                print(f"{player1} won the game")
                break

    elif player2_turn:
           print(f"its {player2}'s turn")
           input("Press enter to roll the dice")
           dice_value = random.choice(dice)
           print(f"{player2} got {dice_value}")
           p2_new_position = player2_pos + dice_value

           if p2_new_position > 100:
               print("oopps....try again in next turn")
           elif p2_new_position in snakes:
               print("Oops! Snake Bite ")
               p2_new_position = snakes[p2_new_position]
               print(f"your new position is {p2_new_position}")
           elif p2_new_position in ladders:
               print("wow! You got a ladder")
               p2_new_position = ladders[p2_new_position]
               print(f"your new position is {p2_new_position}")
           else:
               player2_pos = p2_new_position
               print(f"your new position is {player2_pos}")
               if player2_pos == 100:
                  print(f"{player2} won the game")
                  break
    player1_turn = not player1_turn
    player2_turn = not player2_turn
    input()




```
